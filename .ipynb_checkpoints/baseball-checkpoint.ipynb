## ジオコーディング

https://www.gis-py.com/entry/geopy
https://ohke.hateblo.jp/entry/2018/03/02/230000

```python
def geo(place):
    geolocator = Nominatim(user_agent="my-application")
    location = geolocator.geocode(place)
    
    if location:
        return location.latitude, location.longitude 
    else:
        return "error"
```

## 経度・緯度から2点間距離を計算

https://qiita.com/damyarou/items/9cb633e844c78307134a

```python
def cal_phi(ra,rb,lat):
    return atan(rb/ra*tan(lat))


def cal_rho(lat_a,lon_a,lat_b,lon_b):
    ra=6378.140  # equatorial radius (km)
    rb=6356.755  # polar radius (km)
    F=(ra-rb)/ra # flattening of the earth
    rad_lat_a=radians(lat_a)
    rad_lon_a=radians(lon_a)
    rad_lat_b=radians(lat_b)
    rad_lon_b=radians(lon_b)
    pa=cal_phi(ra,rb,rad_lat_a)
    pb=cal_phi(ra,rb,rad_lat_b)
    xx=acos(sin(pa)*sin(pb)+cos(pa)*cos(pb)*cos(rad_lon_a-rad_lon_b))
    c1=(sin(xx)-xx)*(sin(pa)+sin(pb))**2/cos(xx/2)**2
    c2=(sin(xx)+xx)*(sin(pa)-sin(pb))**2/sin(xx/2)**2
    dr=F/8*(c1-c2)
    rho=ra*(xx+dr)
    return rho


def dis(a_pos, b_pos):
    lat_a,lon_a = a_pos[1], a_pos[0] 
    lat_b,lon_b = b_pos[1], b_pos[0] 
    
    rho=cal_rho(lat_a,lon_a,lat_b,lon_b)
    #print('Distance={0:10.3f} km'.format(rho))
    return rho
```

```python
distans = {
    t2 : { t1 : dis(team_pos[t1], team_pos[t2]) for t1 in team_pos if t1 != t2} for t2 in team_pos
}
```

```python
pd_dis = pd.DataFrame(distans)
pd_dis.round(3) #km
```

## 本拠地間距離の平均と観客動員数の関係
#### データ整形 

```python
df1 = t_dat['H合計']
df2= t_dat.T['V合計']
df3 = pd_dis.mean().round(3)

df = pd.concat([df1,df2,df3], axis=1)
df = df.rename(columns={'H合計': 'ホーム観客動員数','V合計': 'ビジター観客動員数',0: '本拠地間距離の平均'})

df['観客動員数'] = df1 + df2

df = df.drop(['H合計', 'V合計', '交流戦'])
```

```python
df.sort_values("本拠地間距離の平均")
```

```python
n_df1 = (t_dat['H合計']/min(t_dat['H合計'])).round(3)
n_df2= (t_dat.T['V合計']/min(t_dat.T['V合計'])).round(3)
n_df3 = (pd_dis.mean()/min(pd_dis.mean())).round(3)

n_df = pd.concat([n_df1,n_df2,n_df3], axis=1)
n_df = n_df.rename(columns={'H合計': 'ホーム観客動員数','V合計': 'ビジター観客動員数',0: '本拠地間距離の平均'})

temp = t_dat['H合計'] + t_dat.T['V合計']
n_df['観客動員数'] = ((t_dat['H合計'] + t_dat.T['V合計'])/min(t_dat['H合計'] + t_dat.T['V合計'])).round(3)

n_df = n_df.drop(['H合計', 'V合計', '交流戦'])
```
